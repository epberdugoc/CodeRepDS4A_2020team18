```python
num_rows = 0
for exam in ('Pro', 'TyT'):
    for section in ('Generic',):
        for filename in filenames[exam][section]:
            print(filename)
            df = pd.read_csv(filename, dtype=dtypes)
            print(df.shape)
            num_rows += df.shape[0]
            print()
        
print('Total: ', num_rows, 'rows')
```

```python
def get_df(exam, section):
    df_merged = pd.DataFrame()
    mapping_to_rename = {
        'APLI_CONSECUTIVO': 'ESTU_CONSECUTIVO'
    }

    for filename in filenames[exam][section]:
        df = pd.read_csv(filename, dtype=dtypes)
        df['filename'] = filename
        df['exam'] = exam
        df['section'] = section
        df.columns = map(str.upper, df.columns)
        df = df.rename(columns=mapping_to_rename)
        df_merged = pd.concat([df_merged, df])
        
    return df_merged
```

```python
df_pro_generic = get_df('Pro', 'Generic')
df_pro_specific = get_df('Pro', 'Specific')
df_tyt_generic = get_df('TyT', 'Generic')
df_tyt_specific = get_df('TyT', 'Specific')
```

```python
df_merged = pd.concat([df_pro_generic, df_tyt_generic])
```
