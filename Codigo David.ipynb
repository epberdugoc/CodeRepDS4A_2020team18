# AJUSTE DISTRIBUCIONAL

```python
a=datset['MOD_RAZONA_CUANTITAT_PUNT'].dropna()
b=datset['MOD_LECTURA_CRITICA_PUNT'].dropna()
c=datset['MOD_COMPETEN_CIUDADA_PUNT'].dropna()
d=datset['MOD_INGLES_PUNT'].dropna()
e=datset['MOD_COMUNI_ESCRITA_PUNT'].dropna()
f=datset['MOD_COMUNI_ESCRITA_DESEM'].dropna()

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from tqdm import tqdm
import pandas as pd

DISTRIBUTIONS = [        
        st.alpha,st.anglit,st.arcsine,st.beta,st.betaprime,st.bradford,st.burr,st.cauchy,st.chi,st.chi2,st.cosine,
        st.dgamma,st.dweibull,st.erlang,st.expon,st.exponnorm,st.exponweib,st.exponpow,st.f,st.fatiguelife,st.fisk,
        st.foldcauchy,st.foldnorm,st.frechet_r,st.frechet_l,st.genlogistic,st.genpareto,st.gennorm,st.genexpon,
        st.genextreme,st.gausshyper,st.gamma,st.gengamma,st.genhalflogistic,st.gilbrat,st.gompertz,st.gumbel_r,
        st.gumbel_l,st.halfcauchy,st.halflogistic,st.halfnorm,st.halfgennorm,st.hypsecant,st.invgamma,st.invgauss,
        st.invweibull,st.johnsonsb,st.johnsonsu,st.ksone,st.kstwobign,st.laplace,st.levy,st.levy_l,st.levy_stable,
        st.logistic,st.loggamma,st.loglaplace,st.lognorm,st.lomax,st.maxwell,st.mielke,st.nakagami,st.ncx2,st.ncf,
        st.nct,st.norm,st.pareto,st.pearson3,st.powerlaw,st.powerlognorm,st.powernorm,st.rdist,st.reciprocal,
        st.rayleigh,st.rice,st.recipinvgauss,st.semicircular,st.t,st.triang,st.truncexpon,st.truncnorm,st.tukeylambda,
        st.uniform,st.vonmises,st.vonmises_line,st.wald,st.weibull_min,st.weibull_max,st.wrapcauchy
    ]

def fit_scipy_distributions(array, bins, plot_hist = True, plot_best_fit = True, plot_all_fits = False):
    
    if plot_best_fit or plot_all_fits:
        assert plot_hist, "plot_hist must be True if setting plot_best_fit or plot_all_fits to True"
    
    # Returns un-normalised (i.e. counts) histogram
    y, x = np.histogram(np.array(array), bins=bins)
    
    # Some details about the histogram
    bin_width = x[1]-x[0]
    N = len(array)
    x_mid = (x + np.roll(x, -1))[:-1] / 2.0 # go from bin edges to bin middles
    
    # selection of available distributions
    # CHANGE THIS IF REQUIRED
    DISTRIBUTIONS = [st.alpha,st.cauchy,st.cosine,st.laplace,st.levy,st.levy_l,st.norm]

    if plot_hist:
        fig, ax = plt.subplots()
        h = ax.hist(np.array(array), bins = bins, color = 'w')

    # loop through the distributions and store the sum of squared errors
    # so we know which one eventually will have the best fit
    sses = []
    for dist in tqdm(DISTRIBUTIONS):
        name = dist.__class__.__name__[:-4]

        params = dist.fit(np.array(array))
        arg = params[:-2]
        loc = params[-2]
        scale = params[-1]

        pdf = dist.pdf(x_mid, loc=loc, scale=scale, *arg)
        pdf_scaled = pdf * bin_width * N # to go from pdf back to counts need to un-normalise the pdf

        sse = np.sum((y - pdf_scaled)**2)
        sses.append([sse, name])

        # Not strictly necessary to plot, but pretty patterns
        if plot_all_fits:
            ax.plot(x_mid, pdf_scaled, label = name)
    
    if plot_all_fits:
        plt.legend(loc=1)

    # CHANGE THIS IF REQUIRED
    ax.set_xlabel('x label')
    ax.set_ylabel('y label')

    # Things to return - df of SSE and distribution name, the best distribution and its parameters
    results = pd.DataFrame(sses, columns = ['SSE','distribution']).sort_values(by='SSE') 
    best_name = results.iloc[0]['distribution']
    best_dist = getattr(st, best_name)
    best_params = best_dist.fit(np.array(array))
    
    if plot_best_fit:
        new_x = np.linspace(x_mid[0] - (bin_width * 2), x_mid[-1] + (bin_width * 2), 1000)
        best_pdf = best_dist.pdf(new_x, *best_params[:-2], loc=best_params[-2], scale=best_params[-1])
        best_pdf_scaled = best_pdf * bin_width * N
        ax.plot(new_x, best_pdf_scaled, label = best_name)
        plt.legend(loc=1)
    
    if plot_hist:
        plt.show()
    
    return results, best_name, best_params

# Distribucion para MOD_RAZONA_CUANTITAT_PUNT
sses, best_name, best_params = fit_scipy_distributions(a, bins = 100)
print("a:",*best_params[:-2],"loc:",best_params[-2],"scale:", best_params[-1])
#sses, best_name, best_params = fit_scipy_distributions(y, 100, plot_best_fit=False, plot_all_fits=True)
```

```python
# Estandarizar la variable
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
# Matriz de Covarianza
import numpy as np
media_vector = np.mean(X_std, axis=0)
cov_mat = (X_std - media_vector).T.dot((X_std - media_vector)) / (X_std.shape[0]-1)
print('Matriz de covarianza \n%s' %cov_mat)
```

```python
# Descomposicion en valores propios de matriz de covarianza
cov_mat = np.cov(X_std.T)
valorespropios, vectorespropios = np.linalg.eig(cov_mat)

print('Vectores propios \n%s' %vectorespropios)
print('\nValores propios \n%s' %valorespropios)
```

```python
# Decomposicion segun la matriz de correlacion (da lo mismo que con la matriz de covarianza)
matrizcorrelacion = np.corrcoef(X_std.T)
valorespropios, vectorespropios = np.linalg.eig(matrizcorrelacion)

print('Vectores propios \n%s' %vectorespropios)
print('\nValores propios \n%s' %valorespropios)
```

```python
# Comprobacion con descomposicion en valores singulares SVD
#u,s,v = np.linalg.svd(X_std.T) # Problema con la memoria
```

```python
# Verificar si los vectores propios son unitarios
for ev in vectorespropios:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Los vectores son unitarios . DIVIERTETE')
```

```python
# Creaar una lista con las tuplas (valorpropio, vectorpropio) 
parejas = [(np.abs(valorespropios[i]), vectorespropios[:,i]) for i in range(len(valorespropios))]
# Ordenear las tuplas (valorpropio, vectorpropio) en orden descendente 
parejas.sort()
parejas.reverse()
# Verificat si la lita quedo ordenada bien
print('Valores propios en orden descendente:')
for i in parejas:
    print(i[0])
```

```python
# Seleccion de componentes a usar
tot = sum(valorespropios)
var_exp = [(i / tot)*100 for i in sorted(valorespropios, reverse=True)]
print('Porcentajes de varianza explicados por cada componente: \n%s')
cum_var_exp = np.cumsum(var_exp)
print('\nPorcentajes de varianza explicados por cada componente: \n%s')
var_exp,cum_var_exp
# Se puede ver que con dos ejes es suficiente con alrededor de 84% de la varianza total explicada
```

```python
# Reduccion del espacio de 6 dimensiones en uno de 2 eligiendo los dos vectores propios con mayor valor propio
matrixdavid = np.hstack((parejas[0][1].reshape(6,1), parejas[1][1].reshape(6,1)))

print('Matrix David', matrixdavid)
```

```python
# Poryeccion en el nuevo espacio
Y = X_std.dot(matrixdavid)
len(Y)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x = StandardScaler().fit_transform(X)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df1[['ESTU_GENERO']]], axis = 1)
finalDf.head()
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['F', 'M', '-']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['ESTU_GENERO'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
