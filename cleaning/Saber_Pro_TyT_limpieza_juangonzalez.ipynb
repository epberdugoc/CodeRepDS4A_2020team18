### ESTU_NACIONALIDAD, ESTU_GENERO, ESTU_COD_RESIDE_DEPTO, ESTU_COD_RESIDE_MCPIO

```python
def clean_ESTU_NACIONALIDAD(element):
    split_element = element.split()
    
    if split_element[0].isnumeric():
        return ' '.join(split_element[1:]).upper()
    return element.upper()

def clean_ESTU_GENERO(element):
    if element in ['nan', np.nan, '-']:
        return np.nan
    return element.upper()

def clean_ESTU_COD_RESIDE_DEPTO(element):
    if element in ['nan', np.nan, '-'] or math.isnan(float(element)):
        return np.nan
    try:
        return f"{int(float(element)):02d}"
    except:
        print(element in ['nan', np.nan, '-', None])
        print(element)
        print(type(element))
        raise Exception('sth went wrong')
    

def clean_ESTU_COD_RESIDE_MCPIO(element):
    if element in ['nan', np.nan, '-'] or math.isnan(float(element)):
        return np.nan
    return f"{int(float(element)):05d}"
```

```python
clean_functions = {
    'ESTU_NACIONALIDAD': clean_ESTU_NACIONALIDAD,
    'ESTU_GENERO': clean_ESTU_GENERO,
    'ESTU_COD_RESIDE_DEPTO': clean_ESTU_COD_RESIDE_DEPTO,
    'ESTU_COD_RESIDE_MCPIO': clean_ESTU_COD_RESIDE_MCPIO
}

for column, function in clean_functions.items():
    df[column] = df[column].apply(function)
```

```python
def clean_ESTU_NUCLEO_PREGRADO(element):
    #print(element)
    if str(element) in ['nan']:
        return element
    
    split_element = element.split()
    if split_element == []:
        return element
    elif split_element[0].isnumeric():
        return ' '.join(split_element[1:]).upper()
    return element.upper()
```

```python
df['ESTU_NUCLEO_PREGRADO'] = df['ESTU_NUCLEO_PREGRADO'].apply(clean_ESTU_NUCLEO_PREGRADO)
```
